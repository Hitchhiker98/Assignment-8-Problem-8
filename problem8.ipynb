```
import matplotlib.pyplot as plt
import numpy as np

# Step 2: Generate 100 random x and y coordinates
x = np.random.rand(100)
y = np.random.rand(100)

# Step 3: Create a scatter plot
plt.scatter(x, y)

# Step 4: Label the axes
plt.xlabel("X")
plt.ylabel("Y")

# Step 5: Display the plot
plt.show()
```
